```python
df_fin.loc[:,['ITEM TYPE']]
```

```python
#grouping and aggregate-'groupby'

df_fin.groupby('WAREHOUSE SALES').sum()
```

```python
#total amount spent by per category
df_fin.groupby('ITEM TYPE')['RETAIL SALES'].sum()
```

```python
# calculating 4 types metrics using groupby agg()
df_fin.groupby('ITEM TYPE')['RETAIL SALES'].agg(['sum', 'mean', 'max', 'count'])
```

#**INTERPRETATION OF AGG() CALCULATION.**
The results provide valuable insights into the sales performance across different item types. 'LIQUOR' and 'WINE' are the top-performing categories in terms of total retail sales, with 'LIQUOR' also showing a high average sale amount. Conversely, 'DUNNAGE' and 'KEGS' appear to have no retail sales, suggesting they might be related to inventory management rather than direct consumer purchases. 'NON-ALCOHOL' items, while having lower total sales, exhibit the highest average sale per transaction, indicating higher individual item values.
